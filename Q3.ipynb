## Q3) b

* Given $F_{A0} = 0.2mol/s$
* Volume will be the area under $X$ vs $\frac{-F_{A0}}{r_A}$
* From below calculation, we see that the volume would be $0.773\ m^3$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([0, 0.1, 0.2, 0.4, 0.6, 0.7, 0.8])
F = 0.2
rA = -1 * np.array([0.45, 0.37, 0.3, 0.195, 0.113, 0.079, 0.05])
y = -F / rA

plt.plot(X, y, 'o-k')
plt.xlabel('X')
plt.ylabel('$\\frac{-F_{A0}}{r_A}$')

x = X[0:6]
y = y[0:6]
plt.fill_between(x, 0, y)
area = np.trapz(y, x=x)
print("Area under curve:", area)
```
